```python
df_monthly = pd.read_csv('../data/dc_bikes_new.csv', index_col=0, parse_dates=["datetime"])
df_monthly
```

```python
df_monthly["rolling"] = df_monthly["count"].rolling(window=6).mean()
```

```python
plt.plot(df_monthly.index, df_monthly["count"])
plt.plot(df_monthly.index, df_monthly["rolling"])
plt.xticks(rotation=38)
plt.title("Data with rolling average/monthly")
plt.show();
```

```python
df["week_day"] =  df["datetime"].dt.day_name()
```

```python
df.head()
```

```python
df_week_days_cas = df.pivot_table(
    values=["casual", "registered", "count"],
    columns=["week_day"],
    aggfunc="mean",
    sort=False
)
df_week_days_cas = df_week_days_cas.transpose()
```

```python
df_week_days_cas
```

```python
plt.figure(figsize=(10,6))
sns.set_style("darkgrid")
plt.title("Casual, Registred, Count")
sns.lineplot(data=df_week_days_cas, y="casual", x=df_week_days_cas.index, label="Casual")
sns.lineplot(data=df_week_days_cas, y="registered", x=df_week_days_cas.index, label="Registred")
sns.lineplot(data=df_week_days_cas, y="count", x=df_week_days_cas.index, label="Count");
```

```python
df_pivot_hour_day = df.pivot_table(
    values="count",
    index=["hour"],
    columns=["week_day"],
    aggfunc="sum",
    sort=False
)
df_pivot_hour_day 
```

```python
plt.figure(figsize=(10,6))
sns.set_style("darkgrid")
plt.title("Hour per day")
sns.lineplot(data=df_pivot_hour_day, y="Saturday", x="hour", label="Saturday")
sns.lineplot(data=df_pivot_hour_day, y="Sunday", x="hour", label="Sunady")
sns.lineplot(data=df_pivot_hour_day, y="Monday", x="hour", label="Monday")
sns.lineplot(data=df_pivot_hour_day, y="Tuesday", x="hour", label="Tuesday")
sns.lineplot(data=df_pivot_hour_day, y="Wednesday", x="hour", label="Wednesday")
sns.lineplot(data=df_pivot_hour_day, y="Thursday", x="hour", label="Thursday")
sns.lineplot(data=df_pivot_hour_day, y="Friday", x="hour", label="Friday")
plt.xticks(rotation=45);
```

```python
df_2011 = pd.read_csv('../data/dc_bikes_new.csv', index_col=0, parse_dates=['datetime'])
df_2011 = df_2011.set_index("datetime")
df_2011
```

```python
df_2011 = df_2011.loc["2011-05"]
```

```python
df_may_11 = df_2011.groupby("day")["count"].sum().to_frame().reset_index()
df_may_11 
```

```python
df_may_11
```

```python
df_2012 = pd.read_csv('../data/dc_bikes_new.csv', index_col=0, parse_dates=['datetime'])
df_2012 = df_2012.set_index("datetime")
df_2012
```

```python
df_2012=df_2012.loc["2012-05"]
```

```python
df_may_12 = df_2012.groupby("day")["count"].sum().to_frame().reset_index()
df_may_12 
```

```python
plt.figure(figsize=(10,6))
sns.set_style("darkgrid")
plt.title("Count per day 'May 2011, May 2012'")
sns.lineplot(x="day", y="count", data=df_may_12)
sns.lineplot(x="day", y="count", data=df_may_11)
plt.text(0, df_may_12["count"].mean()+10, "Mean of forecast count", color='green')
plt.axhline(df_may_12["count"].mean(), color='green', ls='--')
plt.legend(["forecasts", "actual"]);
```

```python
import statsmodels.tools.eval_measures as sm
```

```python
df_rmse_count = sm.rmse(df_may_11["count"], df_may_12["count"])
df_rmse_count
```
